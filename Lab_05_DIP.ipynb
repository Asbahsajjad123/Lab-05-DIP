```python
# ================== Import Libraries ==================
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from scipy.ndimage import convolve, median_filter, generic_filter, gaussian_filter

# ================== Upload Image at Runtime ==================
uploaded = files.upload()              # Upload image when notebook runs
image_path = list(uploaded.keys())[0]  # Get uploaded image name

# ================== Display Original Image ==================
try:
    img = Image.open(image_path)
    plt.imshow(img)
    plt.axis('off')
    plt.title(f'Displayed Image: {image_path}')
    plt.show()
except Exception as e:
    print("Error displaying image:", e)

# ================== Mean Filter ==================
try:
    img_gray = Image.open(image_path).convert('L')
    img_array = np.array(img_gray)

    plt.imshow(img_array, cmap='gray')
    plt.axis('off')
    plt.title('Original Image (Grayscale)')
    plt.show()

    kernel_size = 3
    mean_filter = np.ones((kernel_size, kernel_size)) / (kernel_size * kernel_size)

    mean_filtered = convolve(img_array, mean_filter, mode='reflect')

    plt.imshow(mean_filtered, cmap='gray')
    plt.axis('off')
    plt.title('Mean Filtered Image')
    plt.show()

except Exception as e:
    print("Error in Mean Filter:", e)

# ================== Median Filter ==================
try:
    median_filtered = median_filter(img_array, size=3)

    plt.imshow(median_filtered, cmap='gray')
    plt.axis('off')
    plt.title('Median Filtered Image')
    plt.show()

except Exception as e:
    print("Error in Median Filter:", e)

# ================== Mode Filter ==================
def mode(arr):
    values, counts = np.unique(arr, return_counts=True)
    return values[np.argmax(counts)]

try:
    mode_filtered = generic_filter(img_array, mode, size=3)

    plt.imshow(mode_filtered, cmap='gray')
    plt.axis('off')
    plt.title('Mode Filtered Image')
    plt.show()

except Exception as e:
    print("Error in Mode Filter:", e)

# ================== Gaussian Filter ==================
try:
    gaussian_filtered = gaussian_filter(img_array, sigma=1)

    plt.imshow(gaussian_filtered, cmap='gray')
    plt.axis('off')
    plt.title('Gaussian Filtered Image')
    plt.show()

except Exception as e:
    print("Error in Gaussian Filter:", e)

# ================== All Filters Together ==================
try:
    plt.figure(figsize=(12, 8))

    plt.subplot(2, 2, 1)
    plt.imshow(img_array, cmap='gray')
    plt.axis('off')
    plt.title('Original Image')

    plt.subplot(2, 2, 2)
    plt.imshow(median_filtered, cmap='gray')
    plt.axis('off')
    plt.title('Median Filter')

    plt.subplot(2, 2, 3)
    plt.imshow(mode_filtered, cmap='gray')
    plt.axis('off')
    plt.title('Mode Filter')

    plt.subplot(2, 2, 4)
    plt.imshow(gaussian_filtered, cmap='gray')
    plt.axis('off')
    plt.title('Gaussian Filter')

    plt.tight_layout()
    plt.show()

except Exception as e:
    print("Error displaying combined results:", e)
```
